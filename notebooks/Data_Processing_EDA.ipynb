Persiapan dan Import Library

```python
#Import Library
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import matplotlib.ticker as ticker

# Pengaturan tampilan agar tabel lebih rapi
pd.set_option('display.max_columns', None)
pd.set_option('display.width', 120)
sns.set(style="whitegrid")
```

Baca Semua Dataset

```python
# Membaca Semua Dataset
coins = {
    "Bitcoin": "coin_Bitcoin.csv",
    "Ethereum": "coin_Ethereum.csv",
    "BinanceCoin": "coin_BinanceCoin.csv",
    "Cardano": "coin_Cardano.csv",
    "Polkadot": "coin_Polkadot.csv"
}

# Membaca dan menyimpan ke dictionary DataFrame
dataframes = {}
for name, path in coins.items():
    df = pd.read_csv(path)
    df['Date'] = pd.to_datetime(df['Date'])
    dataframes[name] = df
    print(f"{name} dataset loaded: {df.shape[0]} rows, {df.shape[1]} columns")
```

Struktur Data dan Informasi Awal

```python
# Menampilkan Struktur Data, Duplikat, dan Cek Outlier
for name, df in dataframes.items():
    print("="*60)
    print(f"{name} Dataset Info")
    print("="*60)
    display(df.head(3))  # tampilkan 3 baris awal

    print("\n--- Info ---")
    df.info()

    print("\n--- Missing Values ---")
    print(df.isnull().sum())

    print("\n--- Data Duplikat ---")
    print(f"Total baris duplikat: {df.duplicated().sum()}")

    print("\n--- Ringkasan Statistik (Cek Outlier) ---")
    display(df.describe())
    print("\n\n")
```

Visualisasi Outlier

```python
# Cek Outlier untuk 3 Kolom (Close, Volume, Marketcap)
print("Mengecek Outlier pada Kolom 'Close', 'Volume', dan 'Marketcap'\n")
fig, axes = plt.subplots(nrows=len(dataframes), ncols=3, figsize=(18, 15))
fig.tight_layout(pad=5.0)

# Loop untuk setiap koin
for i, (name, df) in enumerate(dataframes.items()):

    # Plot 1: Close (di kolom 0)
    sns.boxplot(ax=axes[i, 0], x=df['Close'])
    axes[i, 0].set_title(f"{name} - Close", fontsize=14)
    axes[i, 0].set_xlabel('')

    # Plot 2: Volume (di kolom 1)
    sns.boxplot(ax=axes[i, 1], x=df['Volume'])
    axes[i, 1].set_title(f"{name} - Volume", fontsize=14)
    axes[i, 1].set_xlabel('')

    # Plot 3: Marketcap (di kolom 2)
    sns.boxplot(ax=axes[i, 2], x=df['Marketcap'])
    axes[i, 2].set_title(f"{name} - Marketcap", fontsize=14)
    axes[i, 2].set_xlabel('')

plt.show()
```

Menggabungkan Dataset

```python
# List untuk menampung dataframe yang sudah diberi nama
all_dataframes = []

for name, df in dataframes.items():
    temp_df = df.copy() # Membuat copy agar tidak mengubah dataframe asli di dictionary
    temp_df['CoinName'] = name # Menambahkan kolom 'CoinName' untuk identifikasi di dashboard
    all_dataframes.append(temp_df) # Menambahkan ke listt

all_coins_df = pd.concat(all_dataframes, ignore_index=True) # Menggabubgkan semua dataframe dalam list menjadi satu

print("Semua dataset telah digabungkan!")
print(f"Total baris di dataset gabungan: {all_coins_df.shape[0]}")
print(f"Total kolom di dataset gabungan: {all_coins_df.shape[1]}")

print("\n--- Info Dataset Gabungan ---")
all_coins_df.info()

print("\n--- Contoh Data Gabungan (Acak) ---")
display(all_coins_df.sample(5)) # Menampilkan 5 baris acak dari data gabungan
```

Membersihkan Dataset Gabungan

```python
if 'SNo' in all_coins_df.columns:
    all_coins_df = all_coins_df.drop(columns=['SNo']) # Mengahapus kolom 'SNo' karena tidak lagi menjadi unique value setelah digabung
    print("Kolom 'SNo' telah dihapus (tidak relevan setelah digabung).")
else:
    print("Kolom 'SNo' tidak ditemukan, tidak ada yang dihapus.")

# Tampilkan 5 baris teratas dari data yang sudah bersih
print("\n--- Data Final (Contoh) ---")
display(all_coins_df.head(5))
```

```python
# Memeriksa Ulang & Pembersihan Final Dataset Gabungan
print("="*60)
print("Pengecekan Ulang & Pembersihan Final Dataset Gabungan")
print("="*60)

# Pengecekan & Penanganan Data Duplikat
print("\n1. Data Duplikat")
total_duplikat = all_coins_df.duplicated().sum()
print(f"Total baris duplikat ditemukan: {total_duplikat}")

if total_duplikat > 0:
    all_coins_df.drop_duplicates(inplace=True)
    print("Data duplikat telah dihapus.")

# Pengecekan & Penanganan Missing Values
print("\n2. Missing Values")
print("Missing values (sebelum ditangani):")
missing_values = all_coins_df.isnull().sum() # Hanya menampilkan kolom yang punya missing values
print(missing_values[missing_values > 0])

# Mengisi 'Volume' dan 'Marketcap' dengan 0 (supaya representasi datanya paling jujur dan menjaga visualisasi tetap utuh)
all_coins_df['Volume'] = all_coins_df['Volume'].fillna(0)
all_coins_df['Marketcap'] = all_coins_df['Marketcap'].fillna(0)
print("\nMissing values untuk 'Volume' & 'Marketcap' telah diisi dengan 0.")

# Pengecekan Outlier (via Statistik)
print("\n3. Ringkasan Statistik Final (Cek Outlier)")
print("Statistik ini adalah gambaran data final yang siap diekspor.")
display(all_coins_df.describe())
```

Cek Rentang Waktu

```python
# Rentang Waktu Data (dari Data Gabungan)
print("Rentang Waktu Data per Koin (dari Data Gabungan):")

# Grupkan (pakai kolom 'CoinName', ambil min dan max lalu hitung jumlahnya), agregasi, dan tampilkan hasilnya
stats_waktu = all_coins_df.groupby('CoinName')['Date'].agg(
    Tanggal_Mulai='min',
    Tanggal_Akhir='max',
    Total_Hari='count'
)
display(stats_waktu)

# Rentang Waktu Keseluruhan (lebih ringkas)
print("\nRentang Waktu Keseluruhan (Semua Koin):")
print(f"{all_coins_df['Date'].min().date()}  {all_coins_df['Date'].max().date()}")
```

Visualisasi Tren Harga Penutupan

```python
plt.figure(figsize=(16, 8))
ax = sns.lineplot(
    data=all_coins_df,
    x='Date',
    y='Close',
    hue='CoinName'
)

ax.set_yscale('log')
formatter = ticker.FuncFormatter(lambda y, pos: f'${y:g}')
ax.yaxis.set_major_formatter(formatter)
ax.set_title('Tren Harga Penutupan Koin (Skala Logaritmik)', fontsize=16)
ax.set_xlabel('Tanggal', fontsize=12)
ax.set_ylabel('Harga Penutupan (USD)', fontsize=12)

plt.legend(title='Koin')
plt.grid(True, which="both", ls="--", alpha=0.5)
plt.show()
```

Visualisasi Volume Perdagangan

```python
plt.figure(figsize=(16, 8))
ax = sns.lineplot(
    data=all_coins_df,
    x='Date',
    y='Volume',
    hue='CoinName'
)
ax.set_yscale('log')
ax.set_title('Tren Volume Transaksi (Skala Logaritmik)', fontsize=16)
ax.set_xlabel('Tanggal', fontsize=12)
ax.set_ylabel('Jumlah Transaksi (Skala Log)', fontsize=12)

plt.legend(title='Koin')
plt.grid(True, which="both", ls="--", alpha=0.5)
plt.show()
```

Cek Korelasi Antarkolom

```python
fig, axes = plt.subplots(nrows=1, ncols=len(dataframes), figsize=(28, 6))
fig.suptitle('Matriks Korelasi Antar Kolom (per Koin)', fontsize=18, y=1.05)

for i, (name, df) in enumerate(dataframes.items()):
    sns.heatmap(df[['Open','High','Low','Close','Volume','Marketcap']].corr(), ax=axes[i], annot=True, fmt='.2f', cmap='coolwarm', annot_kws={'size': 9})
    axes[i].set_title(f"{name}", fontsize=14)

plt.tight_layout(rect=[0, 0, 1, 0.95])
plt.show()
```

Ekspor Data Gabungan

```python
# Ekspor ke file CSV
output_filename = "all_currencies_data.csv"
all_coins_df.to_csv(output_filename, index=False)

print(f"Dataset gabungan telah disimpan sebagai '{output_filename}'")
```
